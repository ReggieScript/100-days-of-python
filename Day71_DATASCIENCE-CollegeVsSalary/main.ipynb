What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

```python
max_loc = clean_df["Mid-Career Median Salary"].idxmax()

clean_df["Undergraduate Major"][max_loc]
```

```python
salary = clean_df.loc[max_loc]["Mid-Career Median Salary"]

print(f"graduates from this career earn {salary} at their mid-career stage")
```

Which college major has the lowest starting salary and how much do graduates earn after university?

```python
min_loc = clean_df["Starting Median Salary"].idxmin()

minStartingDegree = clean_df["Undergraduate Major"][min_loc]
```

```python
min_salary = clean_df.loc[min_loc]["Starting Median Salary"]

print(f"graduates from the {minStartingDegree} degree earn {min_salary} after university")
```

```python
clean_df[clean_df["Undergraduate Major"] == minStartingDegree]
```

Which college major has the lowest mid-career salary and how much can people expect to earn with this degree? 

```python
min_min_loc =clean_df["Mid-Career Median Salary"].idxmin()

minMidCareerDegree = clean_df["Undergraduate Major"][min_min_loc]
```

```python
clean_df[clean_df["Undergraduate Major"] == minMidCareerDegree]
```

Lowest Risk Majors

```python
low_risk = clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
clean_df.insert(1, 'Low risk', low_risk)
clean_df.head()
```

```python
lowest_risk_loc = clean_df["Low risk"].idxmin()

lowestRiskDegree = clean_df["Undergraduate Major"][lowest_risk_loc]

clean_df[clean_df["Undergraduate Major"] == lowestRiskDegree]
```

```python
highest_risk_loc = clean_df["Low risk"].idxmax()

highestRiskDegree = clean_df["Undergraduate Major"][highest_risk_loc]

clean_df[clean_df["Undergraduate Major"] == highestRiskDegree]
```

```python
clean_df.sort_values(by = ["Low risk"], ascending=True).head()
```

```python
clean_df.sort_values(by = ["Low risk"], ascending=False).head()
```

```python
clean_df.sort_values(by = ["Mid-Career Median Salary"], ascending= False).head()
```
